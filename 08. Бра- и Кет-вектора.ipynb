
Имеется два кубита, которые могут иметь 2 состояния — 0 и 1. Таким образом, число из этих двух кубитов может иметь $2^2=4$ состояния:

$<\phi> = a_0<00> + a_1<01> + a_2<10> + a_3<11>$

$a$ обозначают **вероятности нахождения этого числа из кубитов в конкретно взятом состоянии**. При короткой записи часть в векторе (<>) может не записываться — тогда они идут в порядке увеличения значения самого числа. (так, $a_0$ соответствует <00>, $a_1$ соответствует <01> и так далее).

ДО ИЗМЕНЕНИЯ система находится в состоянии суперпозиции (как бы сразу во всех состояниях). ПОСЛЕ ИЗМЕНЕНИЯ произойдёт т.н. схлопывание (редукция Фон-Неймана) и кубиты, в некоторой вероятностью, перейдут в какое-то из состояний.

*Подавая, например, лазерные импульсы, мы можем менять условия протекания системы, и таким образом влиять на вероятности состояний, увеличивая их или уменьшая (но об этом уже поговорим на лекции).*

Когда мы измеряем НЕ ВСЕ КУБИТЫ, задача усложняется — нам могут удовлетворять сразу несколько состояний. При этом, поскольку мы остальные кубиты И НЕ ИЗМЕРЯЛИ, он дальше продолжает находиться в состоянии суперпозиции. В записи мы можем вынести известный кубит, и далее для вероятностей записывать только остальные кубиты:

$a_0<00> + a_1<01> \iff <0>(a_0<0> + a_1<1>)$

$a_2<10> + a_3<11> \iff <1>(a_2<0> + a_3<1>)$

Если записать с **нормировочными коэффициентами** (нужен, чтобы сумма вероятснотей всегда была равна 1):

$(k_1 a_0)^2 + (k_1 a_1)^2 = 1$, где $(k_1 a_0)$ соответствует <00>, а $(k_1 a_1)$ соответствует <01>.

Зная вероятности, мы можем найти нормировочный коэффициент:

$k_1 = \frac{1}{\sqrt{a_0^2 + a_1^2}}$

# № 1

В бинарном числе, состоящем из трёх кубитов:

Первая цифра (кубит) — <1>

Вторая цифра (кубит) — <0> <1>

Третья цифра (кубит) — <0> <1>

$<\phi> = \alpha <0> + \beta <1>$

$a_0 = 0.3, a_2 = 0.1, a_2 = 0.9, a_3 = 0.3$

Помним, что вероятность перехода систему в состояние — **это коэф. этого состояния в квадрате**, при этом сумма этих вероятностей должна быть, очевидно, **равна 1** (так как система должна перейти в какое-то одно из этих состояний с вероятностью $100 \%$):

100 >>> $a_0^2 = 0.09$

101 >>> $a_1^2 = 0.01$

110 >>> $a_2^2 = 0.09$

111 >>> $a_3^2 = 0.81$

$\sum = 1.00$

# №2

В бинарном числе, состоящем из трёх кубитов:

$<\phi> = a_0<000> + a_1<001> + a_2<010> + a_3<100> + a_4<011> + a_5<101> + a_6<110> + a_7<111>$

Вероятности:

$a_0 = 0.3, a_1 = -0.6, a_2 = -0.1, a_3 = 0, a_4 = -0.7, a_5 = 0.1, a_6 = -0.7, a_7 = 0$

*(минус это фаза, но не та фаза, которая в электричестве, а фаза, о которой вы узнаете на лекции)*

Нам нужно изменить только первые два кубита, третий нас не интересует. То есть первые два кубита известны, а третий нет.

$\frac{1}{\sqrt{a_0^2+a_1^2}}$

*(нормировочный кожффициент нужен, чтобы суммарно была единица)*

100 >>> $a_0<0> + a_1<1> \implies P = a_0^2 + a_1^2 = 0.45$

101 >>> $a_2<0> + a_4<1> \implies P = 0.5$

110 >>> $<1> \implies P = 0.01$

111 >>> $<0> \implies P = 0.04$

# №3

Имеется состояние $<\phi> = \alpha<0> + \beta<1>$.

Вероятность $\beta = 0.9$, найти вероятность $\alpha$:

$P(\beta) = 0.9 \implies$ $\beta = 0.81$

$\alpha = 1 - \beta = 1 - 0.81 = 0.19$

$\implies P(\alpha) = \sqrt{0.19}$.

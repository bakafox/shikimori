# Интерферометр Маха-Цендера

![](res/09.2.png)

…работает по протоколу **BB84** на **фазовом кодировании** (на позиционном он реализуется примерно так же, но 0/1 задаются не абсолютным наличием/отсутстсвия сигнала, а сдвигом волны по фазе$*$). Это значит, что:

Влетает фотон, на первом зеркале расщепляется, затем отражается от двух зеркал и вылетает на выход.

В зависимости от внесения какой-то фазовой задержки, фотон поменяет своё направление вылета:

Мы можем это детектить, поставив, например, два детектора на выхода. Если вылетает с одной стороны, это 0, если с другой, это 1.

Таким образом, кодироваться 0 и 1 у нас будут фазовым сдвигом на $\frac14$ длины волны.

$*$ _Здесь фаза — это сдвиг по длине волны $\lambda = 2*\pi$. Если представить sin/cos как единичную окружность, то сдвиг на "половину" этого круга будет составлять $\pi$. Фазовый сдвиг $\frac\pi2$, при этом, не меняет самого сигнала (0/1), но прпятствует интерференции одного фотона с другим._

Мы можем продолжить эту схему таким образом:

![](res/09.3.png)

Это и есть **двойной интерферометр Маха-Цендера**. _Для тупых: квадратики — это НЕ РЕЗИСТОРЫ!!! а пружинки — НЕ КАТУШКИ ИНДУКТИВНОСТИ!!!!_

Пошёл по долгому пути _(где типа пружинка)_ — пришёл попозже, пошёл по крóткому — пришёл пораньше.

Тогда, на графике распределения **числа прилётов фотона во времени** снизу:
- Слева будут фотоны, прошедшие по короткому пути (поэтому самая меньшая задержка)
- Справа будут фотоны, прошедшие по длинному пути (поэтому самая бóльшая задержка)
- Посередине — фотоны, которые интерферировали сами с собой (то есть прошли ЛИБО по короткому ЛИБО по длинному пути, мы не знаем по какому). Для последних эта схема работает точно так же, как и самый обычный ~~советский копеешный~~ Интерферометр Маха-Цендера:

Поляризация _(обозначена стрелочками)_ нужна, чтобы исключить фотоны, которые оба раза пройдут по одному и тому же виду пути. Например, фотон проходит ТОЛЬКО по длинному пути, поляризуется вертикально, и после этого проходит по короткому пути.

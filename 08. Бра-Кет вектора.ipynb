# Кет-вектора и Бра-вектора

***Кет-вектор*** $| \ \psi>$ (произносится «кет пси») — это вектор-состояние.
- Пример: $| \ 0>,| \ 1010>, ...$
- С помощью кет-векторов записывается вероятность состояний кубитов: $| \ \psi> \ = \alpha| \ 0 > + \ \beta| \ 1 >$
- В векторе из нескольких кубитов (напр., $| \ 101 >$) кубиты идут слева направо (1-й, 2-й, 3-й...)

***Бра-вектор*** — это «развёрнутый» кет.
- Если $| \ \psi>$ — это столбец (колоночный вектор)…
- …то $<\psi \ |$ — это строка (его эрмитово-сопряжённый вид).
- Встречается гораздо реже кет-векторов.

***Зачем это нужно?*** Чтобы можно было писать:

- Внутреннее произведение: $<\phi \ | \ \psi>$ (даёт число, например вероятность)
- Внешнее произведение (операторы): $| \ \phi><\psi \ |$ (даёт матрицу)
- И всё квантовое вычисление построено на таких операциях.

---

Имеется два кубита, они могут иметь 2 состояния — 0 и 1. Таким образом, число из этих двух кубитов может иметь $2^2=4$ состояния:

$$| \ \phi> \ = a_0| \ 00> + \ a_1| \ 01> + \ a_2| \ 10> + \ a_3| \ 11>,$$

где $a_i$ обозначают **амплитуды вероятностей нахождения этого числа из кубитов в конкретно взятом состоянии**. При короткой записи часть в векторе может не записываться — тогда они идут в порядке увеличения значения самого числа. (так, $a_0$ соответствует $| \ 00>$, $a_1$ соответствует $| \ 01>$ и так далее).

Возведённая в квадрат амплитуда будет означать саму, собственно, вероятность: $P_0 = a_o^2$

---

ДО ИЗМЕНЕНИЯ система находится в состоянии суперпозиции (как бы сразу во всех состояниях). ПОСЛЕ ИЗМЕНЕНИЯ произойдёт т.н. схлопывание (редукция Фон-Неймана) и кубиты, ***с некоторой вероятностью***, перейдут в какое-то из этих состояний.

*Подавая, например, лазерные импульсы, мы можем менять условия протекания системы, и таким образом влиять на вероятности состояний, увеличивая их или уменьшая (но об этом уже поговорим на лекции).*

---

Когда мы измеряем НЕ ВСЕ КУБИТЫ, задача усложняется — **нам могут удовлетворять сразу несколько состояний**. При этом, поскольку мы остальные кубиты И НЕ ИЗМЕРЯЛИ, он дальше продолжает находиться в состоянии суперпозиции. В записи мы можем вынести известный кубит, и далее для вероятностей записывать только остальные кубиты:

$a_0| \ 00> + \ a_1| \ 01> \iff | \ 0>(a_0| \ 0> + \ a_1| \ 1>)$

$a_2| \ 10> + \ a_3| \ 11> \iff | \ 1>(a_2| \ 0> + \ a_3| \ 1>)$

***Мы группируем состояния по значению измеренного кубита.***

---

Мы можем записать с **нормировочными коэффициентами** (это просто число, которое «доводит» сумму вероятностей группы до 1):

$$(k_1 a_0)^2 + (k_1 a_1)^2 = 1,$$

где $(k_1 a_0)$ соответствует $| \ 00>$, а $(k_1 a_1)$ соответствует $| \ 01>$.

Зная вероятности, мы можем найти нормировочный коэффициент: $k_1 = \frac{1}{\sqrt{a_0^2 + a_1^2}}$

# № 1

В бинарном числе, состоящем из трёх кубитов:

- Первая цифра (кубит) — $| \ 1>$

- Вторая цифра (кубит) — $| \ 0> | \ 1>$

- Третья цифра (кубит) — $| \ 0> | \ 1>$

<!-- $| \ \phi> \ = \alpha | \ 0> + \ \beta | \ 1>$ -->

Амплитуды вероятностей:

$a_0 = 0.3, a_1 = 0.1, a_2 = 0.9, a_3 = 0.3$

Найти все вероятности.

---

Поскольку 1-я цифра всегда равна 1, её мы не учитываем.

Помним, что вероятность перехода систему в состояние — **это коэф. этого состояния в квадрате**, при этом сумма этих вероятностей должна быть, очевидно, **равна 1** (так как система должна перейти в какое-то одно из этих состояний с вероятностью $100 \%$):

$P| \ 100> \ = a_0^2 = 0.09$

$P| \ 101> \ = a_1^2 = 0.01$

$P| \ 110> \ = a_2^2 = 0.09$

$P| \ 111> \ = a_3^2 = 0.81$

$\sum = 1.00$

# №2

В бинарном числе, состоящем из трёх кубитов:

$| \ \phi> \ = a_0| \ 000> + \ a_1| \ 001> + \ a_2| \ 010> + \ a_3| \ 011> + \ a_4| \ 100> + \ a_5| \ 101> + \ a_6| \ 110> + \ a_7| \ 111>$

Амплитуды вероятностей:

$a_0 = 0.3, a_1 = -0.6, a_2 = -0.1, a_3 = 0, a_4 = -0.7, a_5 = 0.1, a_6 = -0.7, a_7 = 0$

*(минус это фаза, но не та фаза, которая в электричестве, а фаза, о которой вы узнаете на лекции)*

Измерить [нормированные] вероятности первых двух кубитов (третий нас не интересует).

---

*($x$ в начале означает, что там либо 1, либо 0 — нам не важно)*

$P| \ x00> = \ | \ 0> + \ | \ 0> \implies P = a_0^2 + a_4^2 = 0.09 + 0.49 = 0.58$

$P| \ x01> = \ | \ 0> + \ | \ 1> \implies P = a_1^2 + a_5^2 = 0.36 + 0.01 = 0.37$

$P| \ x10> = \ | \ 1> + \ | \ 0> \implies P = a_2^2 + a_6^2 = 0.01 + 0.49 = 0.50$

$P| \ x11> = \ | \ 1> + \ | \ 1> \implies P = a_3^2 + a_7^2 = 0 + 0 = 0$

Вероятности нашли, теперь выполняем **нормирование вероятностей**, чтобы не учитывать амплитуду вероятности третьего кубика *(т.е. чтобы после сложения всех вероятностей была единица, сейчас же получается чуть ли ни полтора!)*:

$\sum{P} = 0.58 + 0.37 + 0.50 + 0 = 1.45$

NORM $P| \ x00> = \frac{0.58}{1.45} ≈ 0.4$

NORM $P| \ x01> = \frac{0.37}{1.45} ≈ 0.26$

NORM $P| \ x10> = \frac{0.50}{1.45} ≈ 0.34$

NORM $P| \ x11> = \frac{0.37}{1.45} ≈ 0$

Проверим: NORM $\sum{P} = 0.4 + 0.26 + 0.34 + 0 = 1$.

***На этом всё!***

---

Ради интереса найдём также все вероятностые коэффициенты ($\large{\frac{1}{\sqrt{(\sum{a_i^2}})} = \frac{1}{\sqrt{P_i}}}$):

$k| \ x00> = \frac1{\sqrt{0.58}} ≈ 1.3131$

$k| \ x01> = \frac1{\sqrt{0.37}} ≈ 1.6440$

$k| \ x10> = \frac1{\sqrt{0.50}} ≈ 1.4142$

$k| \ x11> = \frac1{\sqrt{0}} ≈ (не \ существует)$

Теперь мы можем применить их к оригинальным амплитудам и получить **нормированные амплитуды** (не путать с **нормированными вероятностями** выше):

NORM $| \ x00 > = 1.3131(0.3 | \ 0> - \ 0.7 | \ 1>) ≈ 0.394 | \ 0 > - 0.919 | \ 1>$

NORM $| \ x01 > = 1.6440(-0.6 | \ 0> + \ 0.1 | \ 1>) ≈ -0.986 | \ 0 > + 0.164 | \ 1>$

NORM $| \ x10 > = 1.4142(-0.1 | \ 0> - \ 0.7 | \ 1>) ≈ -0.141 | \ 0 > - 0.990 | \ 1>$

NORM $| \ x11 > = (не \ существует)$

***Окей круто, но что это всё вообще означает???***

- Нормированные амплитуды — это как «новые веса» первого кубита после того, как мы узнали значение двух последних кубитов.

- Они говорят, насколько вероятно увидеть $| \ 0>$ или $| \ 1>$ для первого кубита для каждого из исходов измерений прошлых двух кубитов.

Рассмотрим на примере NORM $| \ x00 >$:

$P(0|x00) = |0.394| * 2 ≈ 0.155, P(1|x00) = |-0.919| * 2 ≈ 0.845$

Эти числа внутри группы всегда суммируются в 1 (по определению нормировки).

# №3

Имеется состояние $| \ \phi> = \ \alpha| \ 0> + \ \beta| \ 1>$.

Вероятность $P(\beta) = 0.9$, найти вероятность $P(\alpha)$.

---

$P(\beta) = 0.9 \implies$ $\beta = 0.81$

$\alpha = 1 - \beta = 1 - 0.81 = 0.19$

$\implies P(\alpha) = \sqrt{0.19}$.

# №5

Измерить все кубиты для амплитуд вероятностей состояний бинарного числа из трёх кубитов:

$a_0 = 0.3, a_1 = 0.6, a_2 = 0.1, a_3 = 0.6, a_4 = 0.1, a_5 = 0.2, a_6 = 0.2, a_7 = 0.3$

---

***Что значит "измерить все кубиты"?***

Помним, что один кубит записывается как $| \ \psi> \ = \alpha| \ 0 > + \ \beta| \ 1 >$ — мы перечисляем два его состояния.

Тогда несложно догадаться, что три кубита будут записыватья как $\alpha| \ 000 > + \ \beta| \ 001 > + \ \gamma| \ 010 >  + \ \delta| \ 010 > + \ ...$

**Значит, нужно всего лишь найти вероятности каждого из состояний системы.** Помним, что, если не сказано иного, амплитуды вероятностей идут в порядке возрастания самого числа. Тогда, вероятность каждого состояния:

$P| \ 000> \ = a_0^2 = 0.09$

$P| \ 001> \ = a_1^2 = 0.36$

$P| \ 010> \ = a_2^2 = 0.01$

$P| \ 011> \ = a_3^2 = 0.36$

$P| \ 100> \ = a_4^2 = 0.01$

$P| \ 101> \ = a_5^2 = 0.04$

$P| \ 110> \ = a_6^2 = 0.04$

$P| \ 111> \ = a_7^2 = 0.09$

# №6

Измерить последние 2 кубита для вероятностей состояний бинарного числа из трёх кубитов:

$a_0 = 0.3, a_1 = 0.6, a_2 = 0.1, a_3 = 0.1, a_4 = 0.0, a_5 = 0.2, a_6 = 0.2, a_7 = 0.3$

---

Тут всё аналогично задаче №2.

Находим вероятности:

$P| \ x00 > = 0.3^2 + 0.0^2 = 0.09$

$P| \ x01 > = 0.6^2 + 0.2^2 = 0.4$

$P| \ x10 > = 0.1^2 + 0.2^2 = 0.05$

$P| \ x11 > = 0.1^2 + 0.3^2 = 0.1$

Находим их сумму:

$\sum{P} = 0.09 + 0.4 + 0.05 + 0.1 = 0.64$

Нормируем:

NORM $P| \ x00 > = \frac{0.09}{0.64} ≈ 0.14$

NORM $P| \ x01 >  = \frac{0.4}{0.64} ≈ 0.62$

NORM $P| \ x10 > = \frac{0.05}{0.64} ≈ 0.08$

NORM $P| \ x11 > = \frac{0.05}{0.64} ≈ 0.16$

Проверяем сумму:

NORM $\sum{P} = 0.14 + 0.62 + 0.08 + 0.16 = 1$

Сперва разобьём сообщение на тройки (которым сообщение и зашифровано), и определим, какие и насколько часто сочетания встречаются:

```python
i = 0

chars = []
charsdef = {}

while i < len(msg):
    char = msg[i:i+3]

    if not charsdef.get(char):
        charsdef[char] = {
            'count': 1,
            'meaning': f'<{char}>',
        }
    else:
        charsdef[char]['count'] += 1

    chars.append(char)
    i += 3

print(i / 3, len(chars))
print(sorted(charsdef.items(), key=lambda char: char[1]['count']))

# Т(?:\s<\d{3}>){3}\sТ(?:\s<\d{3}>){3}\sТ(?:\s<\d{3}>){3}
# <020> <012> Т <102> <101> <002> Т <112> <000> <202> Т <202> <121> <020>
```

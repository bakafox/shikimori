# Учём суть$^{_{^{_{^{TM}}}}}$?

Представь двух запутанных фотонов (или кубитов), которых отправили Алисe и Бобу. Каждый выбирает направление измерения поляризации:

- Алиса выбирает угол $A$
- Боб выбирает угол $B$

Поскольку фотоны находятся во взаимодействии в одной системе, их поляризации будут ортогональны друг другу.

![](./res/10.1.png)

Если мы возьмём эти фотоны, отправим их две разные лаобтарии — неважно на какое расстояние! — и даже если проведём с ним какие-то эксперименты (дополнительно поляризуем, например) и затем измерим один из фотонов, мы сможем узнать его поляризацию. И, что самое интересное, **у второго фотона ПОСЛЕ этого поляризация при измерении ВСЕГДА будет ортогональна первому** (т.е. если у 1-го она вертикальная, у 2-го она будет горизонатальная, и т.д.).

Как же так, получается обновление происходит быстрее скорости света? На самом деле это не совсем так, и, например, общаться на сверхсветовой скорости таким образом не получится (увы!)

**Кажется, что если как‑то подготовить фотон так, чтобы результат перестал быть случайным, но квантовая теория прямо это запрещает:** Если у Боба фотон — часть запутанной пары, то доля H/V всегда 50/50 (или 50/50 в другой базе, если повернуть поляризатор). И это не зависит от того, что делает Алиса.

# Белл сформулировал неравенство

Пусть у нас есть запутанные кубиты… вспоминаем про состояния Белла, вот все возможные состояния Белла:

$\phi^{\pm} = \frac1{\sqrt2} (<00> \pm <11>)$ 

$\psi^{\pm} = \frac1{\sqrt2} (<01> \pm <10>)$

Таким образом, система будет выглядеть так:

$$\phi^+ = \frac1{\sqrt2}(<\rarr \rarr> + <\uarr \uarr>),$$

где $\rarr \iff 0$, $\uarr \iff 1$.

Пример с Алисой и Бобом в правой части — что такое A B C см. в левом части

![](./res/10.2.png)

1) Вероятность того, что Алиса и Боб выбрали один и тот же базис — $\frac13$

2) Вероятность того, что Алиса и Боб не угадали — соответственно, $\frac23$

$\phi = \frac{\sqrt3}2<\searrow> + \frac12<\swarrow> \Rarr \dots$

Посчитаем вероятности всех исходов, что установленная и выбранная выбранная полярность будет одинаковой и у Алисы, и у Боба:

![](./res/10.3.png)

Таким образом, классическая механика говорит, что вероятность получить такое совпадение не может превышать $\frac12$ (здесь оная везде получилось $\frac59$)…
